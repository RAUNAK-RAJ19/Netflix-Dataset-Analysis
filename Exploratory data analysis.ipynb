#univariate analysis

```python
df_movies = df[df['type'] == 'Movie'].copy()
df_movies['duration_int'] = df_movies['duration'].str.replace(' min', '').astype(int)
df_tv = df[df['type'] == 'TV Show'].copy()
df_tv['duration_int'] = df_tv['duration'].str.replace(' Season', '').str.replace('s', '').astype(int)
```

```python
type_counts = df['type'].value_counts()
```

```python
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values, color=['Blue', 'salmon'])
plt.title('Distribution of content types')
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
movies_count= df_movies['duration_int'].value_counts().sort_index()
```

```python
plt.figure(figsize=(6,4))
plt.plot(movies_count.index, movies_count.values)
plt.title('Duration of movies')
plt.xlabel('duration')
plt.ylabel('count')
plt.show()
```

```python
rating_count=df['rating'].value_counts()
print(rating_count)
```

```python
plt.figure(figsize=(10,6))
plt.bar(rating_count.index, rating_count.values, color='lime', edgecolor='black')
plt.title('Histogram of Netflix Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

```
